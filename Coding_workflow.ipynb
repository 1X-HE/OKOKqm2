**IMD**：

In evaluating the link between Airbnb density and crime rates in London, the importance of neighbourhood socio‑economic context as a potential confounding factor has been recognised. The UK Index of Multiple Deprivation (IMD 2019) offers a comprehensive and detailed theoretical framework for this purpose. However, its direct inclusion in the regression models of this study is subject to two key methodological constraints. First, the composite IMD index itself includes a “crime deprivation” domain. Introducing the overall IMD score as a control variable in a model where crime rate is the dependent variable would create serious endogeneity issues, leading to biased estimates. Second, the IMD data are released at the Lower‑layer Super Output Area (LSOA) level, whereas the key variables in this study—Airbnb density and others—are aggregated to the London borough level. Directly using the raw IMD data would thus introduce a mismatch in the scale of analysis.

To address these limitations and to better disentangle the pathways through which different structural factors may influence crime rates, this study adopts the following strategy. We obtained average scores for each of the seven deprivation domains of the IMD. In the analysis, these domain scores are treated as separate covariates and tested individually in the models. This approach serves a dual purpose: first, it allows us to control for key socio‑economic dimensions such as income, employment, and education while avoiding the circular explanation that would arise from including a crime‑containing composite index; second, the method enables to show potentially distinct association patterns between different dimensions of social deprivation and crime rate.
